```python
import matplotlib.pyplot as plt
import math

%matplotlib inline

plt.style.use("default")

angulo = 30
cos =  math.cos(math.radians(angulo))
sen = math.sin(math.radians(angulo))

def circulo(cos, sen):
  eixos = plt.subplot()

  eixos.grid(True)

  eixos.set_aspect(1)
  eixos.add_artist(plt.Circle((0, 0), 1, fill = False))

  eixos.axhline(y = 0, color= 'black')
  eixos.axvline(x = 0, color= 'black')

  eixos.plot([0, cos], [0, 0],  color = 'green', label = 'cosseno')
  eixos.plot([0, cos], [0, sen],  color = 'red', label = 'hipotenusa')
  eixos.plot([0, 0], [0, sen],  color = 'blue', label = 'seno')
  eixos.plot([0, 1/cos], [1/sen,0], color='yellow', label='tangente')

  plt.xlim(-1.5, 1.5)
  plt.ylim(-1.5, 1.5)

  plt.legend()
  plt.title("Círculo Trigonométrico")
  plt.show()

circulo(cos, sen)
```
