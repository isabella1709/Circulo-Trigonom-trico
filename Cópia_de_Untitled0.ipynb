```python
import matplotlib.pyplot as plt
import math

%matplotlib inline

plt.style.use("default")

def circulo(x, y):
  eixos = plt.subplot()

  eixos.grid(True)

  eixos.set_aspect(1)
  eixos.add_artist(plt.Circle((0, 0), 1, fill = False))

  eixos.axhline(y = 0, color= 'black')
  eixos.axvline(x = 0, color= 'black')

  eixos.plot([0, x], [0, 0],  color = 'green', label = 'cosseno')
  eixos.plot([0, 1], [0, y],  color = 'red', label = 'hipotenusa')
  eixos.plot([0, 0], [0, y],  color = 'blue', label = 'seno')
  eixos.plot([1, 1], [-1, 1], color = 'black', label = 'tangente geral')
  eixos.plot([1, 1], [0.5, 0], color = 'yellow', label = 'tangente')


  plt.xlim(-1.1, 1.1)
  plt.ylim(-1.1, 1.1)

  plt.legend()
  plt.title("Círculo Trigonométrico")
  plt.show()

angulo = 30
x =  math.cos(math.radians(angulo))
y = math.sin(math.radians(angulo))

circulo(x, y)

```
